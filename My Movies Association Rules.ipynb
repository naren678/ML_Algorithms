```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
movie_data=pd.read_csv("C:\\Users\\User\\Downloads\\my_movies.csv")
movie_data.head()
```

```python
movie_data.drop(["V1"],inplace=True,axis = 1)
movie_data.drop(["V2"],inplace=True,axis = 1)
movie_data.drop(["V3"],inplace=True,axis = 1)
movie_data.drop(["V4"],inplace=True,axis = 1)
movie_data.drop(["V5"],inplace=True,axis = 1)
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_itemsets = apriori(movie_data,min_support=0.005,use_colnames=True)
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=1)
rules.head()
```
